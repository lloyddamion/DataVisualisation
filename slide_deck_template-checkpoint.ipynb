## Introduction

<br>

###### Prosper was founded in 2005 as the first peer-to-peer lending marketplace in the United States. Since then, Prosper has facilitated more than \$16 billion in loans to more than 980,000 people.

###### Through Prosper, people can invest in each other in a way that is financially and socially rewarding. Borrowers apply online for a fixed-rate, fixed-term loan between \$2,000 and $40,000. 
###### Individuals and institutions can invest in the loans and earn attractive returns. Prosper handles all loan servicing on behalf of the matched borrowers and investors.

<br>

###### What Is The Minimum Criteria To Borrow On Prosper?

<br>

###### After a borrower submits a loan application, Prosper first obtains a credit report from TransUnion to evaluate whether the applicant meets the 
###### underwriting criteria Prosper has established in conjunction with WebBank. Applicants are subject to the following minimum eligibility criteria:

<br>

 - ###### Debt‐to‐income ratio below 50%
 - ###### Stated income greater than $0
 - ###### No bankruptcies filed within the last 12 months
 - ###### Fewer than five credit bureau inquiries within the last 6 months
 - ###### Minimum of three open trades reported on their credit report
<br>

###### Repeat borrowers must meet several additional criteria:

<br>


 - ###### No previous loans on the Prosper platform which have been charged‐off

 - ###### Must not have been declined for a loan through Prosper within the last four months due to delinquency or returned payments on a previous loan through Prosper

<br>

## Investigation Overview

<br>

###### This project analyzes the personal loan payment dataset to better understand whether the borrower will be able to re-pay the loan.

###### The dataset covers an extensive amount of information on the borrower's side. By further segmenting the loan dataset into defaulted and chargedoff cases and current outstanding loans, this project breaks down the composition of the default cases and examines the correlation among indicators. In the end, the goal is to provide Prosper sufficient evidence on the borrower's ability to repay the loan.

## Dataset Overview
<br>

###### The Prosper dataset contains four kinds of variable categories during 2005 to 2014:

###### * Loan Status : The status of the loan list such as Cancelled, Chargedoff, Completed, Current, Defaulted, FinalPaymentInProgress, PastDue.

###### * Borrower Data : Basic properties about borrowers such as credit score, occupation, employment status, etc.

###### * Loan Data : Basic properties about the loan such as length of the loan(term), Borrower APR, etc.

###### * Credit Risk Matrices : Matrices measured the risk of loans, such as Credit grade, Prosper Score, bank card utilization, etc.

```python
prorat_odr = ['AA','A','B','C','D','E','HR']
proratodr = pd.api.types.CategoricalDtype(ordered = True, categories = prorat_odr)
proloan_nn['ProsperRating (Alpha)'] = proloan_nn['ProsperRating (Alpha)'].astype(proratodr)
```

```python
qtr_odr = ['Q3 2009','Q4 2009','Q1 2010','Q2 2010','Q3 2010','Q4 2010','Q1 2011','Q2 2011','Q3 2011','Q4 2011','Q1 2012','Q2 2012','Q3 2012','Q4 2012','Q1 2013','Q2 2013']
qtrodr = pd.api.types.CategoricalDtype(ordered = True, categories = qtr_odr)
proloan_nn['LoanOriginationQuarter'] = proloan_nn['LoanOriginationQuarter'].astype(qtrodr)
```

```python
ln_sts = ['Current','Completed','Chargedoff','Defaulted','FinalPaymentInProgress','Past Due (1-15 days)','Past Due (16-30 days)','Past Due (31-60 days)','Past Due (61-90 days)','Past Due (91-120 days)','Past Due (>120 days)']
lnsts = pd.api.types.CategoricalDtype(ordered = True, categories = ln_sts)
proloan_nn['LoanStatus'] = proloan_nn['LoanStatus'].astype(lnsts)
```

```python
In_ran = ['$0', '$1-24,999', '$25,000-49,999', '$50,000-74,999', '$75,000-99,999', '$100,000+', 'Not employed']
Inran = pd.api.types.CategoricalDtype(ordered = True, categories = In_ran)
proloan_nn['IncomeRange'] = proloan_nn['IncomeRange'].astype(Inran)
```

```python
proloan_nn['CompCharge'] = np.where(proloan_nn['LoanStatus'].isin(['Defaulted','Chargedoff']), 0, np.where(proloan_nn['LoanStatus'].isin(['Completed']), 1, np.NaN))
```

### Relationship between Credit Score, BorrowerAPR and ProsperRating

<br>

###### Prosper policy is that borrowers must have a minimum credit score of 640 to qualify, the average credit score among Prosper borrowers is 710. 
###### For borrowers with lower credit scores (640-660) the interest rates start to get pretty high, up to 31.99%. [lendacademy](https://www.lendacademy.com/five-things-new-p2p-lending-borrowers-should-know/?)

###### What's evident in the graph is that members with excellent Prosper Rating (AA) have the lowest BorrowerAPR. And members with high risk rating (HR) have the highest BorrowerAPR.
###### When it comes to the Credit score however, it is intersting to note that even with a high score (800 and above) there are members with poor Prosper rating (D, E and HR) and also medium to high BorrowerAPR.
<br>

```python
proloan_nn_t = proloan_nn.copy()
proloan_nn_t = proloan_nn_t.loc[proloan_nn_t['IncomeVerifiable'] == True & proloan_nn_t['DebtToIncomeRatio'].notnull()]
proloan_nn_t = proloan_nn_t.loc[proloan_nn_t['DebtToIncomeRatio'] < 1.0]
proloan_nn_t = proloan_nn_t.loc[proloan_nn_t['StatedMonthlyIncome'] < 100000.00]
proloan_nn_t.dropna(subset=['DebtToIncomeRatio'], inplace = True)

plt.figure(figsize=(20,8))
sb.pointplot(data = proloan_nn_t, x = 'CreditScoreRangeLower', y = 'BorrowerAPR', hue = 'ProsperRating (Alpha)', linestyles = "");
plt.title('Relationship between Credit Score, BorrowerAPR and ProsperRating');
plt.xlabel('Credit Score');
plt.ylabel('Borrower Annual Percentage Rating');
sb.despine(left=True)
```
